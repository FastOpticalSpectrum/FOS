```python
import numpy as np
```

```python
wavelengths = np.linspace(300, 500, 200)
wavelengths
```

```python
alpha = 399
```

```python
len(wavelengths)
```

```python
results = np.concatenate((np.ones(100), np.zeros(100)))
```

```python
np.sum(np.where(wavelengths <= alpha,1-(results),results))
```

```python
np.where(wavelengths <= alpha,1-(results),results)
```

```python
near_indicator_list = []
for alpha in wavelengths:
    # print(alpha)
    near_indicator = np.sum(np.where(wavelengths <= alpha,1-results,results))
    near_indicator_list.append(near_indicator)
near_indicator_array = np.array(near_indicator_list)
min_near_indicator = np.min(near_indicator_array)
argmin_near_indicator = wavelengths[np.argmin(near_indicator_array)]
print(str(min_near_indicator))
print(str(argmin_near_indicator))
```
